## NLP

```python
import nltk
nltk.download('wordnet') # to download wordnet not required always
```

```python
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
ps = WordNetLemmatizer()
```

```python
nltk.download('stopwords')
```

```python
stopwords = stopwords.words('english') #stopwords are a,an,the,or dounle word in sentence like my
```

```python
def clean_row(row):

    row = row.lower() #to lowercase all text
    row = re.sub('[^a-zA-Z]',' ',row) #to remove all special character

    token = row.split() #split every word as a single token

    news = [ps.lemmatize(word) for word in token if not word in stopwords] #lemmetize convert into vector
    
    cleanned_news = ' '.join(news)

    return  cleanned_news
```

```python
dataset['text']
```

```python
dataset['text'] = dataset['text'].apply(lambda x : clean_row(x))  #extra space words are removed
```

```python
dataset['text']
```

```python
from sklearn. feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(max_features = 50000, lowercase = False, ngram_range=(1,2))
```
